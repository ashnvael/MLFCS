```python
import pandas as pd 

df = pd.read_csv(r'uniswap_lp_data\merged_uniswap_data_scaled.csv')
df.head()
```

```python
import datetime as dt

def time_to_seconds(t):
    try:
        h, m, s = map(float, t.split(":"))
        return h * 3600 + m * 60 + s
    except:
        return None

time_part = df['timestamp'].str.extract(r'(\d{2}:\d{2}:\d{2}\.\d+)')
df['timestamp'] = 1e9 * time_part.squeeze().apply(time_to_seconds)
```

```python
df
```

```python
pd.to_datetime(df['timestamp'], unit='s', origin='unix')
```
